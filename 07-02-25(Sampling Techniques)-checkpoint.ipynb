***Up and Down Sampling***

```python
import pandas as pd
from sklearn.utils import resample
df=pd.DataFrame({
    'Age':[22,25,27,20,30,35,40,45,50,55,60,65,70],
    'Income':[2000,2500,2700,3200,3500,3800,4000,4200,4300,4500,5000,5500,6000],
    'Class':['High','Low','Low','High','High','Low','High','High','Low','Low','High','High','Low']
})
```

High Class has 7 instances

Low class has 6 instances

```python
#Separate majority and minority classes
df_high=df[df['Class']=='High']
df_low=df[df['Class']=='Low']
```

```python
df_high
```

```python
df_low
```

```python
#DownSample majority class
df_high_downsampled=resample(df_high,replace=False,n_samples=len(df_low),random_state=42)
```

```python
df_high_downsampled
```

```python
#Combine downsampled majority with minority class
df_balanced=pd.concat([df_high_downsampled,df_low])
```

```python
df_balanced
```

```python
print(df_balanced['Class'].value_counts())
```

```python
#UpSample majority class
df_high_upsampled=resample(df_low,replace=True,n_samples=len(df_high),random_state=42)
```

```python
df_high_upsampled
```

```python
#Combine downsampled majority with minority class
df_balanced=pd.concat([df_high_upsampled,df_high])
```

```python
print(df_balanced['Class'].value_counts())
```

***Example-2***

```python
df=pd.DataFrame({
    'Age':[22,25,27,20,30,35,40,45,50,55,60,65,70],
    'Income':[2000,2500,2700,3200,3500,3800,4000,4200,4300,4500,5000,5500,6000],
    'Class':['Minority','Majority','Majority','Majority','Majority','Minority','Minority','Majority','Majority','Majority','Minority','Majority','Majority']
})
```

```python
#Separate majority and minority classes
df_majority=df[df['Class']=='Majority']
df_minority=df[df['Class']=='Minority']
```

```python
df_majority
```

```python
df_minority
```

```python
#DownSample majority class
df_majority_downsampled=resample(df_majority,replace=False,n_samples=len(df_minority),random_state=42)
df_majority_downsampled
```

```python
#Combine downsampled majority with minority class
df_balanced=pd.concat([df_majority_downsampled,df_minority])
df_balanced
```

```python
print(df_balanced['Class'].value_counts())
```

```python
#DownSample majority class
df_minority_upsampled=resample(df_minority,replace=True,n_samples=len(df_majority),random_state=42)
df_minority_upsampled
```

```python
#Combine downsampled majority with minority class
df_balanced=pd.concat([df_minority_upsampled,df_majority])
df_balanced
```

```python
print(df_balanced['Class'].value_counts())
```
